### üéØ Objective
The objective of this project is to analyze customer purchase behavior using effective data cleaning and visualization techniques. Visual insights help in understanding trends, patterns, and relationships within the data.


```python

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## üìÇ Dataset Loading

```python

# Load dataset (CSV file must be in the same folder)"C:\\Users\\Sharwan Kumar\\OneDrive\\Documents\\D 2.csv"")
df=pd.read_csv("C:\\Users\\Sharwan Kumar\\OneDrive\\Documents\\D 2.csv")
df
df.head()
```

## üßπ Data Cleaning

```python

# Check missing values
df.isnull().sum()

# Remove missing values
df.dropna(inplace=True)

# Convert date column
df['Last_Purchase_Date'] = pd.to_datetime(df['Last_Purchase_Date'])

df['Last_Purchase_Date'] = pd.to_datetime(
    df['Last_Purchase_Date'],
    format='%d-%m-%Y'
)

```

## üìà Visualization 1: Age Distribution

```python

plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```


### üîç Insight:
- Most customers belong to a **specific middle-age group**, indicating the primary target audience.
- Very young and very old age groups contribute less to purchases.


## üìâ Visualization 2: Age vs Total Purchases

```python

plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Total_Purchases', data=df)
plt.title("Relationship Between Age and Total Purchases")
plt.xlabel("Age")
plt.ylabel("Total Number of Purchases")
plt.tight_layout()
plt.show()
```


### üîç Insight:
- Purchase behavior shows **moderate variation with age**.
- Middle-aged customers tend to make slightly more purchases.
- There is no strong linear relationship, indicating other influencing factors.


## üì¶ Visualization 3: Distribution of Total Purchases

```python

plt.figure(figsize=(8,5))
sns.boxplot(x=df['Total_Purchases'])
plt.title("Distribution of Total Purchases")
plt.xlabel("Total Number of Purchases")
plt.tight_layout()
plt.show()
```


### üîç Insight:
- Majority of customers make purchases within a limited range.
- A few customers act as **high-frequency buyers**, visible as outliers.


## üõí Visualization 4: Purchase Category Frequency

```python
from collections import Counter

# Flatten categories into a list
categories = df['Purchase_Categories'].dropna().str.split(',').sum()
categories = [c.strip() for c in categories]

# Count frequency
cat_counts = Counter(categories)

plt.figure(figsize=(8,5))
sns.barplot(x=list(cat_counts.keys()),y=list(cat_counts.values()),hue=list(cat_counts.keys()), palette="pastel")
plt.title("Popular Purchase Categories", fontsize=16, fontweight='bold')
plt.xlabel("Category") # labeling
plt.ylabel("Frequency") # labeling
plt.xticks(rotation=45) # rotating the ticks 
plt.show()
```


### üîç Insight:
- Certain categories dominate customer preferences.
- Businesses should focus marketing strategies on **high-frequency categories**.


## ‚è≥ Visualization 5: Purchases Over Time

```python

df_sorted = df.sort_values('Last_Purchase_Date')

plt.figure(figsize=(8,5))
plt.plot(df_sorted['Last_Purchase_Date'], df_sorted['Total_Purchases'])
plt.title("Purchasing Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Purchases")
plt.tight_layout()
plt.show()
```


### üîç Insight:
- Purchase trends show **fluctuations over time**.
- Peaks may correspond to seasonal offers or promotions.


## üîó visualization 6 : Correlation Heatmap



```python
plt.figure(figsize=(8,5))
sns.heatmap(df[['Age', 'Total_Purchases']].corr(), annot=True)
plt.title("Correlation Between Age and Total Purchases")
plt.tight_layout()
plt.show()
```

### üîç Insight:
- The correlation between age and total purchases is weak to moderate.
- This indicates that age alone does not strongly influence purchasing behavior.
- Other factors such as preferences, income, or offers may play a role.
